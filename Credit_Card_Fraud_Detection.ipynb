This Dataset is highly unbalanced

```python
#separating the data for analysis.
legit = card_data[card_data.Class == 0]
fraud = card_data[card_data.Class == 1]
```

Build a sample dataset containing similar distribution of Normal transactions and Fraud transactions.

```python
legit_sample = legit.sample(n = 492)
```

Concatenating two DataFrames

```python
new_dataset = pd.concat([legit_sample, fraud], axis = 0) #in the new dataset, legit_sample and fraud will be in the 'Class' row.

new_dataset.head()
```

```python
new_dataset.shape
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
X = new_dataset.drop(columns = 'Class', axis = 1)
Y = new_dataset['Class']

print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, stratify = Y, random_state = 3)

print(X.shape, X_train.shape, X_test.shape)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
X_train_prediction = model.predict(X_train)

result = accuracy_score(X_train_prediction, Y_train)

print(result)
```

```python
X_test_prediction = model.predict(X_test)

result_test = accuracy_score(X_test_prediction, Y_test)

print(result_test)
```
